```python
import pyranges as pr
import pandas as pd
```

```python
annotation = pr.read_gff3("gene.gff3") 
annotation.head()
```

```python
annotation[['gene_name']].head()
```

```python
import os
import glob
bed_files = glob.glob('*.bed')
for file in bed_files:
    try :
        print(file)
        featname = os.path.splitext(file)[0]  
        print(featname)
        peaks = pr.read_bed(file)
        print(peaks.head(2))
        overlapping_genes = peaks.join(annotation, how="left") 
        print(overlapping_genes.columns)
        
        df_overlapping_genes = overlapping_genes.df
        df=df_overlapping_genes[['gene_name','Score']].drop_duplicates()
        df = df.rename(columns={'gene_name': 'gene'})
        df = df.rename(columns={'Score':featname})
        print(df.head(2))
        df.to_csv(featname+'.csv',index=False)
    except:
        raise
```

```python
import pandas as pd
import os
import glob

csv_files = glob.glob('*.csv')
merged_df = pd.DataFrame()
for i, file in enumerate(csv_files):
    current_df = pd.read_csv(file)
    print(current_df.shape)
    current_df = current_df[~current_df['gene'].str.startswith('ENSG0')].drop_duplicates(subset='gene')
    print(current_df.shape)
    if i == 0:
        merged_df = current_df
    else:
        print("")
        merged_df = pd.merge(merged_df, current_df, on='gene', how='left')  # 'outer' 
        print(i,merged_df.shape)
merged_df.to_csv('BREAST_CHIPSEQ.csv',index=False)
```
